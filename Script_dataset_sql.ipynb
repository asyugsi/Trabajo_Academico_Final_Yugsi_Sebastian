```python
import sqlite3
import pandas as pd
from faker import Faker
import random
```

2. Se crea o abre una base de datos llamada "coffeedb.sqlite", se crea la tabla llamada "cafe" con la definición de las columnas correspondientes.

```python
conn = sqlite3.connect("data/coffeedb.sqlite")
cursor = conn.cursor() #controlador para ejecutar comandos SQL

# Creación de la tabla y definición de las columnas y tipos de datos a ingresarse
cursor.execute(""" 
CREATE TABLE cafe (
    id INTEGER PRIMARY KEY,
    Region TEXT,
    Altitude INTEGER,
    Productor TEXT,
    Price REAL
)
""")

conn.commit()
conn.close()
```

3. Se agregan registros a la tabla de la base de datos haciendo uso de el módulo faker para los nombres de las empresas productoras de café, para la columna "Region" se hizo uso de random.choise para elegir aleatoriamente entre las provincias definidas en la lista "provincia". Así mismo, se hizo uso de random para numeros aleatorios para "Altitude" y para "Price" 

```python
conn = sqlite3.connect("data/coffeedb.sqlite")
cursor = conn.cursor()

fake = Faker('es_ES')

productores = [fake.unique.company() for _ in range(10)]
provincia = ["Manabí", "Orellana", "Loja", "Sucumbios", "Imbabura", "Zamora Chinchipe", "El Oro"]

for i in range(5000):
    region = random.choice(provincia)
    altitude = random.randrange(1200, 1800, 100)
    productor = random.choice(productores)
    price = round(random.uniform(5, 10), 2)

    cursor.execute("""
        INSERT INTO cafe
        (Region, Altitude, Productor, Price)
        VALUES (?, ?, ?, ?)
        """, ((region, altitude, productor, price))
    )

conn.commit()
conn.close()
```

4. Visualización con Pandas de la tabla de base de datos generada, solo los 5 primeros registros.

```python
conn = sqlite3.connect("data/coffeedb.sqlite")
df = pd.read_sql_query("SELECT * FROM cafe", conn)
print(df.head())
conn.close()
```
